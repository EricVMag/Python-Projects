```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://locations.traderjoes.com/ca/los-angeles'

```

```python
def zipcode_extract(url):

  request = requests.get(url)
  request

  source = request.content

  soup = BeautifulSoup(source, 'html.parser')

  address_divs = soup.find_all('div', attrs={'class':'address-left'})

  zipcodes = []

  for div in address_divs:
    zipcode = div.find_all('span')[4].text.strip()
    zipcodes += [zipcode]

  return(zipcodes)
```

```python
def location_urls(url):

  request = requests.get(url)
  request
  source = request.content
  soup = BeautifulSoup(source, 'html.parser')

  locations_div = soup.find('div', attrs={'class':'col-md-12'})
  locations_a = locations_div.find_all('a')

  urls = []

  for a in locations_a:
    urls += [a.get('href')]

  return(urls)
```

```python
def state_urls(url):

  request = requests.get(url)
  request
  source = request.content
  soup = BeautifulSoup(source, 'html.parser')

  states_div = soup.find('div', attrs={'class':'col-md-12'})
  states_a = states_div.find_all('a')

  urls = []

  for a in states_a:
    urls += [a.get('href')]

  return(urls)
```

```python
url = 'http://locations.traderjoes.com/'

zipcodes = []

for state in state_urls(url):
  for loc in location_urls(state):
    zipcode = zipcode_extract(loc)
    print(zipcode)
    zipcodes.extend(zipcode)
```

## Locations Map

```python
from geopy.geocoders import Nominatim
import folium
```

```python
geolocator = Nominatim(user_agent="myGeocoder")

latitudes = []
longitudes = []

for zipcode in zipcodes:
  location = geolocator.geocode(zipcode + ", USA")
  latitudes.append(location.latitude)
  longitudes.append(location.longitude)
  print(str(location.latitude)+ ", " + str(location.longitude))
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

geolocator = Nominatim(user_agent="your_specific_user_agent")

latitudes = []
longitudes = []

for zipcode in zipcodes:
    try:
        location = geolocator.geocode(zipcode + ", USA")
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
        print(str(location.latitude)+ ", " + str(location.longitude))
    except GeocoderTimedOut:
        print("GeocoderTimedOut: Retrying...")
        continue
```

```python
map = folium.Map([39,-98],zoom_start=4)

for lat, lng, zipcode in zip(latitudes, longitudes, zipcodes):
  folium.CircleMarker(location=[lat,lng], radius=1,fill=True,color='#0057B8', popup=zipcode).add_to(map)

map
```
